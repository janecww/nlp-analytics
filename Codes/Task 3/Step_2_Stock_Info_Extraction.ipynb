```python

#https://pypi.org/project/yahoofinancials/
#https://pypi.org/project/yfinance/


def data_extraction(symbol):
  ticker = YahooFinancials(symbol)
  yf_ticker = yf.Ticker(symbol)

  profit_margin = ticker.get_key_statistics_data().get(symbol).get('profitMargins')
  pe_ratio = ticker.get_pe_ratio()
  #print(pe_ratio)
  beta = ticker.get_key_statistics_data().get(symbol).get('beta')
  #print(beta)
  ebita = ticker.get_key_statistics_data().get(symbol).get("enterpriseToEbitda")
  #print(ebita)
  eps = ticker.get_key_statistics_data().get(symbol).get("trailingEps")
  #print(eps)
  #if (yf_ticker.sustainability):
  #RISK SCORES
  total_esg = yf_ticker.sustainability.to_dict().get('Value').get('totalEsg')
    #print(total_esg)
  e_esg = yf_ticker.sustainability.to_dict().get('Value').get('environmentScore')
    #print(e_esg)
  s_esg = yf_ticker.sustainability.to_dict().get('Value').get('socialScore')
    #print(s_esg)
  g_esg = yf_ticker.sustainability.to_dict().get('Value').get('governanceScore')
    #print(g_esg)
  controversy = yf_ticker.sustainability.to_dict().get('Value').get('highestControversy')
    #print(highestControversy)
  #else:
    #total_esg = 0
    #e_esg = 0
    #s_esg = 0
    #g_esg = 0
    #controversy = 0


  df = pd.DataFrame(
      {
          'Symbol':symbol,
          'beta':beta,
          'profit_margin':profit_margin,
          'pe_ratio':pe_ratio,
          'ev/ebita': ebita,
          'eps': eps,
          'total_esg': total_esg,
          'e_esg':e_esg,
          's_esg':s_esg,
          'g_esg':g_esg,
          'highestControversy':controversy

      },index=[symbol]
    )


  #df = df.astype('float')


  return df
```

```python
#for i in tickers:
  #data_extraction(i)
df_all = pd.DataFrame(
       {
            'Symbol':[1],
            'beta':[1],
            'profit_margin':[1],
            'pe_ratio':[1],
            'ev/ebita': [1],
            'eps': [1],
            'total_esg': [1],
            'e_esg':[1],
            's_esg':[1],
            'g_esg':[1],
            'highestControversy':[1]

        }
     )

for ticker in tickers:
  try:
    data = data_extraction(ticker)
    df_all = df_all.append(data, ignore_index=False)
    print(ticker,' --- appended.')
  except:
    print(ticker,' --- An error occurred.')
```

```python
df_all.to_csv('stock_fina_esg_data.csv')
```
