```python
#Extract texts from news only
news = text_df[text_df['pubtype']=='U']
news
```

```python
news.info()
```

```python
#Drop unimportant columns
news = news.drop(columns = ['engagements', 'non_view_engagements', 'emoji_count', 'author_name', 'pubtype', 'region', 'url', 'fans_count'])

#Drop duplicated contents
news = news.drop_duplicates(subset = 'content')
news
```

```python
news.describe()
```

```python
#Fill in N/A values with 0.0
#comment_count	like_count	dislike_count	love_count	haha_count	wow_count	angry_count	sad_count	share_count	view_count
values = {"comment_count": 0.0, "like_count": 0.0, "dislike_count": 0.0, "love_count": 0.0, "haha_count": 0.0, "wow_count":0.0, "angry_count":0.0, "sad_count":0.0, "share_count":0.0, "view_count":0.0}
news = news.fillna(value=values)

# replace -1 iwth 0
news['comment_count'] = news['comment_count'].replace(-1.0, 0.0)
news['like_count'] = news['like_count'].replace(-1.0, 0.0)
news['view_count'] = news['view_count'].replace(-1.0, 0.0)
news
```

```python
news.describe(include = 'all')
```

```python
#Remove all the reaction/engagement counts
news = news.drop(columns = ['docid', 'comment_count', 'like_count', 'dislike_count', 'love_count', 'haha_count', 'wow_count', 'angry_count', 'sad_count', 'share_count', 'view_count'])
```

```python
news
```

```python
news.dropna(axis = 0, how ='any', inplace = True)
news.describe(include = 'all')
```

```python
#Set date and time to be index
news['date'] = news['pubdate'].dt.date
news['time'] = news['pubdate'].dt.time
news = news.set_index(['date', 'time'], 'date')
news = news.sort_values(by= ['date', 'time'])
news
```

```python
# Text
#Remove special characters
news['headline'] = news['headline'].astype(str)
news['content'] =news['content'].astype(str)
news['headline'] = news['headline'].replace('\W', ' ', regex=True)
news['content'] = news['content'].replace('\W', ' ', regex=True)
news
```

```python
headline = news['headline'].to_list()
content = news['content'].to_list()
#Use the join() and split() methods together to remove duplicate spaces and newline characters
for i in range(len(content)):
    content[i] = " ".join(content[i].split())

for i in range(len(headline)):
    headline[i] = " ".join(headline[i].split())

content
```

```python
news['content'] = content
news['headline']= headline
news.head()
```
