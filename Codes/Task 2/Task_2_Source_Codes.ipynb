## K means clustering ##

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# k means clustering version 1
def tokenize_zh(text):
    words = jieba.lcut(text)
    return words

# Add a custom list of stopwords for punctuation
stop_words = ['。', '，',' ','\n','!','\t','"']

vectorizer = CountVectorizer(tokenizer=tokenize_zh, stop_words=stop_words,lowercase=False)
X_train_counts=vectorizer.fit_transform(all_pd_words_sorted_social_copy['content'].astype(str))
```

```python


import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
#Sum_of_squared_distances = []
#K = range(2,10)
#for k in K:
    #km = KMeans(n_clusters=k, max_iter=200, n_init=10)
    #km = km.fit(X_train_counts)
    #Sum_of_squared_distances.append(km.inertia_)
#plt.plot(K, Sum_of_squared_distances, 'bx-')
#plt.xlabel('k')
#plt.ylabel('Sum_of_squared_distances')
#plt.title('Elbow Method For Optimal k')
#plt.show()

print('How many clusters do you want to use?')
true_k = int(input())
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=200, n_init=10)
model.fit(X_train_counts)

labels=model.labels_
clusters=pd.DataFrame(list(zip(all_pd_words_sorted_social_copy['content'],labels)),columns=['title','cluster'])
clusters_copy =clusters.copy()


for i in range(true_k):
    print(clusters[clusters['cluster'] == i])

#print(clusters.sort_values(by=['cluster']))
```
